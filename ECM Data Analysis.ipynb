```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv (r'dat00001.csv')
df = pd.DataFrame(data, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])

data2 = pd.read_csv (r'dat00002.csv')
df2 = pd.DataFrame(data2, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df2)


data3 = pd.read_csv (r'dat00003.csv')
df3 = pd.DataFrame(data3, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df3)

data4 = pd.read_csv (r'dat00004.csv')
df4 = pd.DataFrame(data4, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df4)

data5 = pd.read_csv (r'dat00005.csv')
df5 = pd.DataFrame(data5, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df5)

data6 = pd.read_csv (r'dat00006.csv')
df6 = pd.DataFrame(data6, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df6)

data7 = pd.read_csv (r'dat00007.csv')
df7 = pd.DataFrame(data7, columns = ['Record #','Time','Ch 101 (VDC)','Ch 102 (VDC)','Alarm Output','Signature'])
df = df.append(df7)


plt.figure(figsize = (42,8))
t = df.loc[:,'Record #']/3600
V = df.loc[:,'Ch 101 (VDC)']
I = V/0.015*1000
plt.plot(t, I,'k.')
plt.xlabel('Time (h)')
plt.ylabel('Detected Current (mA)')
axes=plt.gca()
axes.set_xlim([0,120])
plt.savefig('Current_Time.png')
```
